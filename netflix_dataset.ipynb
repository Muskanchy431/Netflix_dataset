```python
from collections import Counter  # Import the Counter class from collections module

country_data = netflix_df['Country'].dropna().apply(lambda x: x.split(', '))
all_countries = []
for countries in country_data:
    all_countries.extend(countries)

country_count = Counter(all_countries)
top_10_countries = pd.DataFrame(country_count.items(), columns=['Country', 'Count']).sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_10_countries, x='Count', y='Country', palette='Blues_d')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```
